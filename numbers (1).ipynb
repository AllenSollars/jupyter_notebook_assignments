```python
#3600 seconds

hours = 1
htm = 60
mts = 60

hours_to_minutes = hours * htm
minute_to_seconds = hours_to_minutes * mts

print(minute_to_seconds)
```

```python
seconds_per_hour = minute_to_seconds
print(seconds_per_hour)
```

```python
#86400 seconds

dth = 24

print(seconds_per_hour * dth)
```

```python
seconds_per_day = seconds_per_hour * dth
print(seconds_per_day)
```

```python
print(seconds_per_day / seconds_per_hour)
```

```python
print(seconds_per_day // seconds_per_hour)

#yes, both answers matched except for the .0
```
