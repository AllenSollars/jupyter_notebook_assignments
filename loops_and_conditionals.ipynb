```python
secret = 5

guess = 3

if secret == guess:
    print("just right")
elif guess < secret:
    print("too low")
else:
    print("too high")
```

```python
small = True
green = False

if small and green:
    print("pea")
elif small and not green:
    print("cherry")
elif not small and green:
    print("watermelon")
elif not small and not green:
    print("pumpkin")
```
