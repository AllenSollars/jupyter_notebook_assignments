```python
from datetime import date

today = date.today()
today_str = str(today)  # e.g. '2026-04-27'

with open('today.txt', 'w') as f:
    f.write(today_str)

print(f"Written to today.txt: {today_str}")
```

```python
with open('today.txt', 'r') as f:
    today_string = f.read()

print(f"today_string = {repr(today_string)}")
```

```python
from datetime import datetime

parsed_date = datetime.strptime(today_string, '%Y-%m-%d').date()

print(f"Parsed date : {parsed_date}")
print(f"Type        : {type(parsed_date)}")
print(f"Year: {parsed_date.year}  Month: {parsed_date.month}  Day: {parsed_date.day}")
```
